```python
import sys
sys.path.append("../source")

import numpy as np

import matplotlib.pyplot as plt

from firefly import FireflyOptimizer
```

```python
def drawPopulation(x, xx, y, yy):
    # Define plot components
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 8))
    
    axes[0].set_xlim([-50, 50])
    axes[0].set_ylim([-50, 50])
    axes[1].set_xlim([-50, 50])
    axes[1].set_ylim([-50, 50])
    
    axes[0].scatter(x[:, 0], x[:, 1], s=3, color="lime")
    axes[0].scatter(y[:, 0], y[:, 1], s=3, color="tomato")
    axes[1].scatter(xx[:, 0], xx[:, 1], s=3, color="lime")
    axes[1].scatter(yy[:, 0], yy[:, 1], s=3, color="tomato")
```

```python
# Sample data
ham = np.random.multivariate_normal([0, 0], [[10, 0], [0, 100]], 100)
spam = np.random.multivariate_normal([15, 15], [[10, 0], [0, 10]], 100)

# Draw populations
#drawPopulation(ham, spam)

model = FireflyOptimizer([-1.e-3, 1.e-3,], maxIter=50)
hham = model.optimize(ham)
sspam = model.optimize(spam)

drawPopulation(ham, hham, spam, sspam)
```
